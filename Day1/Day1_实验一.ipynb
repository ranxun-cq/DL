## 实验信息
1. 数据集：波士顿房价

```python
# 读取数据集
from sklearn.datasets import load_boston
lb = load_boston()
```

```python
# 切分数据集
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(lb.data, lb.target, test_size=0.2)
```

```python
y_train = y_train.reshape(-1,1)
y_test = y_test.reshape(-1,1)
```

```python
# 标准化
from sklearn.preprocessing import StandardScaler
std_x = StandardScaler()
x_train = std_x.fit_transform(x_train)
x_test = std_x.transform(x_test) # 注意与上面句不一样！

std_y = StandardScaler()
y_train= std_y.fit_transform(y_train)
y_test = std_y.transform(y_test)
```

```python
# 建立线性模型
from sklearn.linear_model import LinearRegression, SGDRegressor, RidgeCV
from sklearn.metrics import r2_score
import numpy as np
#########################################################
#                                            线性回归
#########################################################
lr = LinearRegression()
# 拟合
lr.fit(x_train,y_train)
# 预测
y_predict = lr.predict(x_test)
# 评估
r2 = r2_score(y_test, y_predict)
print("r2 score of Linear regression is {}".format(r2))
#########################################################
#                                            岭回归
#########################################################
cv = RidgeCV(alphas=np.logspace(-3,2,100)) # 为什么传入一数列的alpha？
# 拟合
cv.fit(x_train,y_train)
# 预测
y_predict = cv.predict(x_test)
# 评估
r2 = r2_score(y_test, y_predict)
print("r2 score of CVRidge is {}".format(r2))
#########################################################
#                                            SGD回归
#########################################################
sgd = SGDRegressor()
sgd.fit(x_train, y_train)
y_predict = sgd.predict(x_test)
r2 = r2_score(y_test, y_predict)
print("r2 score of SGDRegressor is {}".format(r2))
```

```python
# 建立人工神经网络模型
from keras.models import Sequential 
from keras.layers import Dense
# 构造结构
seq = Sequential()
seq.add(Dense(64, activation='relu', input_dim=lb.data.shape[1])) # 为什么这里要inputdim
seq.add(Dense(64, activation='relu'))
seq.add(Dense(1, activation='relu'))
# 优化配置
seq.compile(optimizer='rmsprop', loss='mse', metrics=['mae']) # metrics参数为什么是列表？
# 拟合
seq.fit(x_train, y_train, batch_size=16, epochs=300, shuffle=False)
# 预测
y_predict = seq.predict(x_test)
# 评估
score = seq.evaluate(x_test, y_test, batch_size = 16) # 什么意思？
print('score is {}'.format(score))
r2 = r2_score(y_test, y_predict)
print("r2 score of MLP is {}".format(r2))
```
